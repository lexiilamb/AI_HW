### Imports

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

## Defining Functions

```
def f1(x,y):
  return ((4 * np.cos(x)**2) + (4 * np.cos(y)))
  
def f2(x,y):
   return ((4 * np.cos(x)**2) + (np.cos(y**2)))
  
def f3(x,y):
   return ((4 * np.cos(x)**2) + (4 * np.cos(y) * np.cos(x)))
```

## Function 1

### 2D Plot

```
# Initialize spacing for points
x = np.linspace(0, 10, 50)
y = np.linspace(0, 10, 50)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

# Plot 2D
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
plt.annotate('local min', xy=(np.pi/2, np.pi), xytext=(np.pi/2, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('local max', xy=(np.pi, 2*np.pi), xytext=(np.pi, 3*np.pi/2), arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('zero grad', xy=(2*np.pi, np.pi), xytext=(2*np.pi, 1.5), arrowprops=dict(facecolor='black', shrink=0.05))
```

### 3D Plot

```
# 3D plot
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

# Adjust view of graph
ax.view_init(30, 30)
```

### Quiver Graph

```
# Modify line spacing
x = np.linspace(0, 15, 30)
y = np.linspace(0, 15, 30)

# Placement of arrows
X, Y = np.meshgrid(x, y)

# Arrow placement 
U = -9*np.sin(X)*np.cos(X)
V = -3*np.sin(Y)

fig, ax = plt.subplots()
q = ax.quiver(X, Y, U, V)

plt.show()
```
